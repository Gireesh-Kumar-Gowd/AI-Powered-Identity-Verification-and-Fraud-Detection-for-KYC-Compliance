## **Draw Results**

```python
import cv2
import matplotlib.pyplot as plt
import openbharatocr

img = cv2.imread(Image_path)

# Extract Voter ID details
front_result = openbharatocr.voter_id_front(Image_path)

font = cv2.FONT_HERSHEY_SIMPLEX
color = (255, 0, 0)
thickness = 2

y = 40
line_gap = 40

for key, value in front_result.items():
    if value:  # skip empty fields
        text = f"{key}: {value}"
        cv2.putText(img, text, (20, y), font, 0.8, color, thickness, cv2.LINE_AA)
        y += line_gap

plt.figure(figsize=(7, 7))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
